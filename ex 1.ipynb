```python
#Histogram 

import matplotlib.pyplot as plt
import numpy as np

# Assuming you have the petal width data (let's call it 'petal_width_data')
petal_width_data = np.random.normal(loc=1.0, scale=0.5, size=100) # Example data

plt.hist(petal_width_data, bins=np.arange(0, 3, 0.2), color='grey', edgecolor='black')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()

#DENSITY PLOT

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have the petal length data (let's call it 'petal_length_data')
petal_length_data = np.random.normal(loc=4.0, scale=0.8, size=100) # Example data

sns.kdeplot(petal_length_data, fill=True)
plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.title('Density Plot of Petal Length')
plt.show()

#Box plot
 
sns.boxplot(x='Species', y='Sepal.Height', data=iris_dataset)
plt.xlabel('Species')
plt.ylabel('Sepal Height')
plt.title('Box Plot of Sepal Height')
plt.show()

#Barchart

# Assuming you have the petal height data (let's call it 'petal_height_data')
petal_height_data = np.random.normal(loc=2.0, scale=0.6, size=100) # Example data

plt.bar(range(len(petal_height_data)), petal_height_data, color='blue')
plt.xlabel('Sample Index')
plt.ylabel('Petal Height')
plt.title('Bar Chart of Petal Height')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv("iris.csv")
df.head()
df.shape
df_seto=df.loc[df["Species"]=="Iris-setosa"]
df_viri=df.loc[df["Species"]=="Iris-virginica"]
df_versi=df.loc[df["Species"]=="Iris-versicolor"]
plt.plot(df_seto["SepalLengthCm"],np.zeros_like(df_seto["SepalLengthCm"]),'o')
plt.plot(df_seto["SepalLengthCm"],np.zeros_like(df_seto["SepalLengthCm"]),'o')
plt.plot(df_viri["SepalLengthCm"],np.zeros_like(df_viri["SepalLengthCm"]),'o')
plt.plot(df_versi["SepalLengthCm"],np.zeros_like(df_versi["SepalLengthCm"]),'o')
#Univariate for sepalwidth

plt.plot(df_seto["SepalWidthCm"],np.zeros_like(df_seto["SepalWidthCm"]),'o')
plt.plot(df_viri["SepalWidthCm"],np.zeros_like(df_viri["SepalWidthCm"]),'o')
plt.plot(df_versi["SepalWidthCm"],np.zeros_like(df_versi["SepalWidthCm"]),'o')
plt.xlabel("sepalwidth")
plt.show()
#univariate for petal width
plt.plot(df_seto["PetalWidthCm"],np.zeros_like(df_seto["PetalWidthCm"]),'o')
plt.plot(df_viri["PetalWidthCm"],np.zeros_like(df_viri["PetalWidthCm"]),'o')
plt.plot(df_versi["PetalWidthCm"],np.zeros_like(df_versi["PetalWidthCm"]),'o')
plt.xlabel("Petalwidth")
plt.show()
#univariate petalLength

plt.plot(df_seto["SepalLengthCm"],np.zeros_like(df_seto["SepalLengthCm"]),'o')
plt.plot(df_viri["SepalLengthCm"],np.zeros_like(df_viri["SepalLengthCm"]),'o')
plt.plot(df_versi["SepalLengthCm"],np.zeros_like(df_versi["SepalLengthCm"]),'o')
plt.xlabel("sepalLength")
plt.show()

#Bivariate sepallength vs petallength

sns.FacetGrid(df,hue="Species").map(plt.scatter,"SepalWidthCm","PetalWidthCm").add_legend()
plt.show()

sns.FacetGrid(df,hue="Species").map(plt.scatter,"SepalLengthCm","PetalLengthCm").add_legend()
plt.show()

#Multivariate
sns.pairplot(df,hue="Species",size=5)
```
