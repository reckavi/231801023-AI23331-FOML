```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression


data = pd.read_csv('suv_data.csv')
X, y = data[['Age', 'EstimatedSalary']], data['Purchased']

model = LogisticRegression().fit(X, y)
y_pred_prob = model.predict_proba(X)[:, 1]


fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for ax, feature, color in zip(axes, ['Age', 'EstimatedSalary'], ['red', 'blue']):
    ax.scatter(data[feature], y, label='Actual')
    ax.scatter(data[feature], y_pred_prob, color=color, alpha=0.5, label='Predicted')
    ax.set_xlabel(feature)
    ax.set_ylabel('Probability of Purchase')
    ax.legend()

plt.tight_layout()
plt.show()


print(f"Intercept: {model.intercept_[0]}, Coefficients: {model.coef_[0]}")
```
